# Section 3. Regression Discontinuity Plots

```r
turki<-read.csv("P:/CODING/GITHUB/econometrics_ugm/Microeconometrics/Regression Discontinuity Design/CIT_2020_CUP_polecon.csv",sep=",", header=1)
head(turki)
```

```r
plot(turki$X, turki$Y,
     xlab = "Score", ylab = "Outcome", main="Scatter Plot (Myersson Data)",
     col = "blue", pch = 20,
     cex.axis = 1.5, cex.lab = 1.5)
abline(v = 0)
# cex = character expansion
# cex.axis = 1.5 -> Mengatur ukuran angka sumbu (angka di sisi kiri dan bawah).
# 1.5 berarti 1.5 kali ukuran default.
# cex.lab = 1.5 -> Mengatur ukuran label sumbu (label "Score" dan "Outcome").
```

## R Snippet 2

```r
library(rdrobust)

out <- rdplot(y = turki$Y,
              x = turki$X,
              nbins = c(20, 20),
              binselect = "es",
              title = "Scatter Plot (Myersson Data)",
              x.label = "Score",
              y.label = "Outcome")

summary(out)
```
